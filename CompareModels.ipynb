```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import numpy as np
```

```python
r = nn.runModel('ETHUSDROC')
```

```python
e = nn.runModel('ETHUSDENS')
```

```python
r.tail(300).SR.prod()
```

```python
e.tail(300).SR.prod()
```

```python
e
```

```python
er = pd.merge(r, e, on='date', suffixes=('_r', '_e'))
er = er[['date', 'close_r','signal_r', 'signal_e', 'SR_r', 'SR_e', 'DR_r']]
er = er.set_index('date')
# er = er.loc['2017':]
er['CSR_r'] = er.SR_r.cumprod()
er['CSR_e'] = er.SR_e.cumprod()
er['CMR'] = er.DR_r.cumprod()

fig, ax = plt.subplots(figsize=(30, 15))
ax.set_yscale('log')
fig.autofmt_xdate()
er['color_r'] = np.where(er.signal_r=='Buy', 'g', 'r')
er['color_e'] = np.where(er.signal_e=='Buy', 'g', 'r')
ax.plot(er['CSR_r'], color='g', label='ROC')
ax.plot(er['CSR_e'], color='b', label='Combo')
ax.plot(er['CMR'], color='r', label='Market Return')
ax.scatter(er.index, er.CSR_r, color=er['color_r'], s=10)
ax.scatter(er.index, er.CSR_e, color=er['color_e'], s=10)
plt.legend()
plt.grid(True)
plt.title('Strategy Performance')
plt.show()
```
